```python
from emm.bin.symmetricize import Symmetricize
import matplotlib.pyplot as plt
import importlib
```

```python
for i in range(100):
    shape = Symmetricize('../emm/matfiles/' + str(i+1) + '.mat')
    
    if(shape.should_rotate("fit")):
        rot_x, rot_y = shape.rotate_figure("fit")       
        rot_then_sym_x, rot_then_sym_y = shape.make_symmetric("rotated", rot_x, rot_y)
        plt.figure(i)
        plt.plot(shape.x_fit, shape.y_fit)
        plt.plot(rot_then_sym_x, rot_then_sym_y)
    else:
        new_x2, new_y2 = shape.make_symmetric("fit")
        plt.figure(i)
        plt.plot(shape.x_fit, shape.y_fit)
        plt.plot(new_x2, new_y2)
        
```

```python
for i in range(100):    
    shape = Symmetricize('../emm/matfiles/' + str(i+1) + '.mat')
    if(shape.should_rotate("outerfit")):
        rot_x, rot_y = shape.rotate_figure("outerfit")       
        rot_then_sym_x, rot_then_sym_y = shape.make_symmetric("rotated", rot_x, rot_y)
        plt.figure(i)
        plt.plot(shape.x_outerfit, shape.y_outerfit)
        plt.plot(rot_then_sym_x, rot_then_sym_y)
    else:
        new_x2, new_y2 = shape.make_symmetric("outerfit")
        plt.figure(i)
        plt.plot(shape.x_outerfit, shape.y_outerfit)
        plt.plot(new_x2, new_y2)
```
